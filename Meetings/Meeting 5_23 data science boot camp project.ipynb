```python
# Tech
tech_tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'NVDA']

# Biotech
biotech_tickers = ['JNJ', 'LLY', 'NVO', 'MRK', 'RHHBY']

# Industrial
industrial_tickers = ['DOW', 'GE', 'F', 'UNP', 'SO']

# Healthcare
healthcare_tickers = ['PFE', 'ABBV', 'CVS', 'AMGN', 'UNH']

# Cryptocurrencies
crypto_tickers = ['BTC', 'ETH', 'BNB', 'XRP', 'ADA']
```

```python
import yfinance as yf
import pandas as pd

# Define the ticker symbols
tech_tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'NVDA']
biotech_tickers = ['JNJ', 'LLY', 'NVO', 'MRK', 'RHHBY']
industrial_tickers = ['DOW', 'GE', 'F', 'UNP', 'SO']
healthcare_tickers = ['PFE', 'ABBV', 'CVS', 'AMGN', 'UNH']
crypto_tickers = ['BTC-USD', 'ETH-USD', 'BNB-USD', 'XRP-USD', 'ADA-USD']

# Define the period of data to retrieve (1 year)
start_date = '2022-01-01'
end_date = '2022-12-31'

#S&P 500
sp500_data = yf.download('^GSPC', start_date, end_date)['Adj Close']
sp500_returns = sp500_data.pct_change()
sp500_volatility = sp500_returns.std()* (252 ** 0.5)

# Load stock data and compute returns for tech stocks
tech_data = yf.download(tech_tickers, start=start_date, end=end_date)['Adj Close']
tech_returns = tech_data.pct_change()
tech_volatility = tech_returns.std() * (252 ** 0.5)  # Assuming 252 trading days in a year

# Load stock data and compute returns for biotech stocks
biotech_data = yf.download(biotech_tickers, start=start_date, end=end_date)['Adj Close']
biotech_returns = biotech_data.pct_change()
biotech_volatility = biotech_returns.std() * (252 ** 0.5)

# Load stock data and compute returns for industrial stocks
industrial_data = yf.download(industrial_tickers, start=start_date, end=end_date)['Adj Close']
industrial_returns = industrial_data.pct_change()
industrial_volatility = industrial_returns.std() * (252 ** 0.5)

# Load stock data and compute returns for healthcare stocks
healthcare_data = yf.download(healthcare_tickers, start=start_date, end=end_date)['Adj Close']
healthcare_returns = healthcare_data.pct_change()
healthcare_volatility = healthcare_returns.std() * (252 ** 0.5)

# Load stock data and compute returns for cryptocurrencies
crypto_data = yf.download(crypto_tickers, start=start_date, end=end_date)['Adj Close']
crypto_returns = crypto_data.pct_change()
crypto_volatility = crypto_returns.std() * (252 ** 0.5)


print("SP500 Volatility (Returns):")
print(sp500_volatility)
print()

# Print the volatility for each category
print("Tech Volatility (Returns):")
print(tech_volatility)
print()

print("Biotech Volatility (Returns):")
print(biotech_volatility)
print()

print("Industrial Volatility (Returns):")
print(industrial_volatility)
print()

print("Healthcare Volatility (Returns):")
print(healthcare_volatility)
print()

print("Crypto Volatility (Returns):")
print(crypto_volatility)
```
